### ```generate_bins``` 
A function that takes the bin size and the dataframe and returns list of numbers that form the start points of bins

```window_size``` = size of the bins (eg. 6000 if you want bins to be \[0, 6000), \[6000, 12000) etc.

```df``` = relevant dataframe

````python
def generate_bins(window_size, df):
    last_index = len(df["POS"]) - 1
    last_element = df["POS"][last_index]
    bins = list(range(0, last_element+window_size, window_size))
    return bins
````

### ```plot_hist``` 
A function that takes the bin size and the dataframe and returns a histogram with the given bins

```window_size``` = size of the bins (eg. 6000 if you want bins to be \[0, 6000), \[6000, 12000) etc.

```df``` = relevant dataframe

````python
def plot_hist(window_size, df):
    bins = generate_bins(window_size, df)
    plt.hist(df["POS"], bins = bins)
    plt.ticklabel_format(style = 'plain')
    plt.xlabel("number of base pairs", fontsize = 50)
    plt.ylabel("number of SNPs", fontsize = 50)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 30)
    #plt.savefig('/Users/anushasubramanian/Desktop/blah.png', dpi = 180, bbox = 'tight')
````

### ```sliding_window``` 
A function that takes the bin size and the dataframe and returns a dataframe with the columns "bins" containing that endpoints of the bins and "data" which has the number of data points in that particular bin.

```window_size``` = size of the bins (eg. 6000 if you want bins to be \[0, 6000), \[6000, 12000) etc.

```df``` = relevant dataframe

````python
# df would be a dataframe and window_size would be an integer

def sliding_window(window_size, df):
    
    bins = generate_bins(window_size, df)
    
    binned_data = []
    for i in range(len(bins)-1):
        count = np.count_nonzero(df["POS"].between(bins[i], bins[i+1]-1, inclusive = True))
        binned_data.append(count)
    to_plot = pd.DataFrame()
    to_plot["bins"] = bins[1:]
    to_plot["data"] = binned_data
    return to_plot
    
````

## SCORING GENOTYPE BY REGION

### SCORING TYPE A

### ```scoring_genotype_1```

````python
def scoring_genotype_1(file):
    """
    function that will take a dataframe and then use a 80% cut off to mark a 
    position as 1/0, 0/1 (homo) or 1/1 (hetero)
    
    """
    
    GT_psl2 = []
    GT_psl5 = []

    for ind in file.index:
        if (file["psl2%"][ind] >= 80) and (file["psl5%"][ind] <= 20):
            GT_psl2.append(1)
            GT_psl5.append(0)
        elif (file["psl2%"][ind] <= 20) and (file["psl5%"][ind] >= 80):
            GT_psl2.append(0)
            GT_psl5.append(1)
        elif (file["psl2%"][ind] < 80) and (file["psl5%"][ind] < 80):
            GT_psl2.append(1)
            GT_psl5.append(1)

    return GT_psl5, GT_psl2
    
````

````python
## run the function on all 7704 files

for key in filesDict:
    
    file = filesDict[key]
    GT_psl5, GT_psl2 = scoring_genotype_1(file)
    
    file["GT psl5 (A)"] = GT_psl5
    file["GT psl2 (A)"] = GT_psl2
    
    filesDict[key] = file
````

### ```GT_percent_psl2```

````python
def GT_percent_psl2(window_size, file, column_name):
    
    """
    window_size = integer for width of bins
    file = dataframe that has the GT psl2 (A) column
    
    """
    bins = generate_bins(6000, file)
    GT_psl2Per = []
    
    for i in range(len(bins)-1):
        
        sliced_df = file.loc[(file["POS"] >= bins[i]) & (file["POS"]< bins[i+1])]
        percent = (sliced_df[column_name].sum()/len(sliced_df)) * 100
        GT_psl2Per.append(percent)
    
    return bins, GT_psl2Per
    
   
````

### ```scoring_genotype_B```

create the GT psl2 (B) column

````python
def scoring_genotype_B(file):
    """
    function that will take a dataframe and then use a 80% cut off to mark a 
    position as 2/0, 0/2 (homo) or 1/1 (hetero)
    
    """
    
    GT_psl2 = []
    GT_psl5 = []

    for ind in file.index:
        if (file["psl2%"][ind] >= 80) and (file["psl5%"][ind] <= 20):
            GT_psl2.append(2)
            GT_psl5.append(0)
        elif (file["psl2%"][ind] <= 20) and (file["psl5%"][ind] >= 80):
            GT_psl2.append(0)
            GT_psl5.append(2)
        elif (file["psl2%"][ind] < 80) and (file["psl5%"][ind] < 80):
            GT_psl2.append(1)
            GT_psl5.append(1)

    return GT_psl5, GT_psl2
    
````

### Calculate indiviudal PSL2%

from the 0,1,2 scored GT psl2

````python
def GT_percent(file):
    
    psl2Per = []   
    
    for GT in file["GT psl2 (B)"]:
        psl2Per.append((GT/2)*100)
    
    return psl2Per
        
````

### ```GT_avg_percent_psl2_B``` for method B

can't use the previous function because now you are averaging percentages

````python
def GT_avg_percent_psl2_B(window_size, file, column_name):
    
    """
    window_size = integer for width of bins
    file = dataframe that has the GT psl2% (B) column
    column_name = GT psl2% (B)  
    
    """
    bins = generate_bins(window_size, file)
    GT_psl2Per = []
    
    for i in range(len(bins)-1):
        
        sliced_df = file.loc[(file["POS"] >= bins[i]) & (file["POS"]< bins[i+1])]
        percent = (sliced_df[column_name].sum()/len(sliced_df))
        GT_psl2Per.append(percent)
    
    return bins, GT_psl2Per
````
