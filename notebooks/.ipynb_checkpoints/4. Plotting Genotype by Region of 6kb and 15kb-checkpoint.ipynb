````python
names = [
"yi283_ACGCGA.CTTTCTCTTTACTTA.csv",
"yi283_TAATGC.CTTTCTCTCTGTTAA.csv",
"yi283_CGGTGG.CTCGATTAACGTCGT.csv",
"yi283_TTCTCC.TACGATATACGTCGT.csv"
]

# only plotting a subset of required plots for example. A lot more plots are present on the drive.
````

````python
psl2Dict = dict.fromkeys(names)
````

````python
len(psl2Dict.keys())
````

````python
# key : value pairing

for filename in psl2Dict:
    psl2Dict[filename] = pd.read_csv(filename)
````

````python
psl2Dict["yi283_ACGCGA.CTTTCTCTTTACTTA.csv"].head() #checking to make sure it works
````

````python
[x for x in psl2Dict.keys() if type(psl2Dict[x]) != pd.core.frame.DataFrame] # check to make all keys have a df 
````

````python
[x for x in psl2Dict.keys() if len(psl2Dict[x]) == 0] # check to see if any df is empty
````

````python
#SET_INDEX
````

````python
sns.set(rc={'figure.figsize':(100,25.27)})
````

````python
#******************** remember to change the axes labels and colorscheme (optional) *********************************

def plot_genotypeA(filesDict):
    
    subset50 = list(filesDict.keys())
    #list(filesDict.keys())[100:150]
    sns.set(rc={'figure.figsize':(100,25.27)})

    for key in subset50:

        GTfile = filesDict[key]
        GTfile.fillna(-1, inplace = True)
        GTfile.set_index("bins", inplace=True)
        GTfilename = key.replace('.csv', '.png')

        plt.figure()
        sns.barplot(x = GTfile.index, 
                y = 'GT avg psl2% (C)', data = GTfile, color='darkorchid')
#darkturquoise
#mediumvioletred
#darkorchid

        plt.title('%PSL2 for sgs1 chrIV positions in 15kb windows', fontsize = 70)
        plt.xticks(fontsize = 20, rotation = 90)
        plt.xlabel('sgs1 positions in 15kb intervals', fontsize = 35)
        plt.yticks(fontsize = 60)
        plt.ylabel('psl2%', fontsize = 50)
        plt.savefig(f'/Users/anushasubramanian/Desktop/(C) GT Plots/{GTfilename}', dpi = 180, bbox_inches = 'tight')
````

````python
plot_genotypeA(psl2Dict)
````
